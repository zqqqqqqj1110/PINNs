```python
import torch
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# 设置GPU设备，如果没有可用的GPU，则使用CPU
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

def setup_seed(seed):
    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True

setup_seed(888888)

print(device)
```

```python
# 基础参数
h = 100    # 画图网格密度
N = 1000    # 内点配置点数
N1 = 0.01    # 边界点配置点数
N2 = 1000    # PDE数据点
```

```python
class MLP(torch.nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.net = torch.nn.Sequential(
            torch.nn.Linear(2, 32),
            torch.nn.Tanh(),
            torch.nn.Linear(32, 32),
            torch.nn.Tanh(),
            torch.nn.Linear(32, 32),
            torch.nn.Tanh(),
            torch.nn.Linear(32, 32),
            torch.nn.Tanh(),
            torch.nn.Linear(32, 1)
        )

    def forward(self, x):
        return self.net(x)
```

```python
# 定义模型并加载权重
u = MLP().to(device)
u.load_state_dict(torch.load('weights.pth', map_location=device))
u.eval()  # 设置模型为评估模式

# 定义第n象限的网格点
h = 500  # 网格的点数，调整以改变分辨率
xc = torch.linspace(0, 1, h, device=device)
yc = torch.linspace(0, 1, h, device=device)
xm, ym = torch.meshgrid(xc, yc)
xx = xm.reshape(-1, 1)
yy = ym.reshape(-1, 1)
xy = torch.cat([xx, yy], dim=1).to(device)

# 使用模型对网格点进行预测
with torch.no_grad():  # 禁用梯度计算，以提高预测速度
    predictions = torch.relu(u(xy))
```

```python
# 计算解析解
analytical_solution = torch.relu(torch.sin(torch.pi * xx) * torch.sin(torch.pi * yy)).to(device)

# 计算绝对误差
absolute_error = torch.abs(predictions - analytical_solution)

# 将预测值、解析解和绝对误差重新整形为网格形状
predictions = predictions.cpu().numpy().reshape(h, h)
analytical_solution = analytical_solution.cpu().numpy().reshape(h, h)
absolute_error = absolute_error.cpu().numpy().reshape(h, h)

# 绘制预测值的3D图形
fig = plt.figure(figsize=(18, 6))

# 预测值图像
ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(xm.cpu().numpy(), ym.cpu().numpy(), predictions, cmap='viridis')
ax1.set_title('Predictions')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_zlabel('Prediction')

# 解析解图像
ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(xm.cpu().numpy(), ym.cpu().numpy(), analytical_solution, cmap='viridis')
ax2.set_title('Analytical Solution')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_zlabel('Solution')

# 绝对误差图像
ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(xm.cpu().numpy(), ym.cpu().numpy(), absolute_error, cmap='viridis')
ax3.set_title('Absolute Error')
ax3.set_xlabel('X')
ax3.set_ylabel('Y')
ax3.set_zlabel('Error')

plt.show()
```

```python
import torch
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# 定义模型
class MLP(torch.nn.Module):
    def __init__(self):
        super(MLP, self).__init__()
        self.net = torch.nn.Sequential(
            torch.nn.Linear(2, 32),
            torch.nn.Tanh(),
            torch.nn.Linear(32, 32),
            torch.nn.Tanh(),
            torch.nn.Linear(32, 32),
            torch.nn.Tanh(),
            torch.nn.Linear(32, 32),
            torch.nn.Tanh(),
            torch.nn.Linear(32, 1)
        )

    def forward(self, x):
        return self.net(x)

# 定义网格大小和其他参数
h = 500  # 网格的点数，调整以改变分辨率
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# 加载和预测函数
def predict_from_weights(weight_file, x_range, y_range):
    u = MLP().to(device)
    u.load_state_dict(torch.load(weight_file, map_location=device))
    u.eval()  # 设置为评估模式
    
    # 创建网格点
    xc = torch.linspace(x_range[0], x_range[1], h, device=device)
    yc = torch.linspace(y_range[0], y_range[1], h, device=device)
    xm, ym = torch.meshgrid(xc, yc, indexing='ij')  # 使用 'ij' 索引方式来确保正确的网格布局
    xx = xm.reshape(-1, 1)
    yy = ym.reshape(-1, 1)
    xy = torch.cat([xx, yy], dim=1).to(device)

    # 使用模型进行预测
    with torch.no_grad():
        predictions = torch.relu(u(xy))
    
    return xm.cpu().numpy(), ym.cpu().numpy(), predictions.cpu().numpy().reshape(h, h)

# 定义四个象限的权重文件和对应的网格范围
weights_files = ['weights1.pth', 'weights2.pth', 'weights3.pth', 'weights4.pth']
ranges = [([0, 1], [0, 1]),   # 第1象限
          ([-1, 0], [0, 1]),  # 第2象限
          ([-1, 0], [-1, 0]), # 第3象限
          ([0, 1], [-1, 0])]  # 第4象限

# 创建一个空的网格，用于存储所有预测结果
predictions_combined = np.zeros((h, h))

# 获取每个象限的预测结果
for i, (weight_file, (x_range, y_range)) in enumerate(zip(weights_files, ranges)):
    # 在每个象限的网格上进行预测
    xm, ym, predictions = predict_from_weights(weight_file, x_range, y_range)
    
    # 根据象限更新预测值
    if i == 0:  # 第1象限
        predictions_combined[:h//2, :h//2] = predictions[:h//2, :h//2]
    elif i == 1:  # 第2象限
        predictions_combined[:h//2, h//2:] = predictions[:h//2, h//2:]
    elif i == 2:  # 第3象限
        predictions_combined[h//2:, h//2:] = predictions[h//2:, h//2:]
    else:  # 第4象限
        predictions_combined[h//2:, :h//2] = predictions[h//2:, :h//2]

# 创建一维张量
xx = torch.linspace(-1, 1, h, device=device)
yy = torch.linspace(-1, 1, h, device=device)

# 使用meshgrid生成网格
xm, ym = torch.meshgrid(xx, yy, indexing='ij')  # 使用 'ij' 索引方式来确保正确的网格布局

# 计算解析解
analytical_solution = torch.relu(torch.sin(torch.pi * xm) * torch.sin(torch.pi * ym)).to(device)

# 将解析解转换为numpy数组
analytical_solution = analytical_solution.cpu().numpy().reshape(h, h)

# 计算误差（绝对误差）
absolute_error = np.abs(predictions_combined - analytical_solution)

# 绘制2D误差图
fig, ax = plt.subplots(figsize=(8, 6))
c = ax.imshow(absolute_error, extent=[-1, 1, -1, 1], origin='lower', cmap='viridis')
fig.colorbar(c, ax=ax)
ax.set_title('Absolute Error (2D)')
ax.set_xlabel('X')
ax.set_ylabel('Y')
```
